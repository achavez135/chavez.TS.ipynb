**Part 1:** *How many good and bad characters are there in total? and how many good and bad characters are in each universe?*

**Description:** *In analyzing the alignment of the characters and what universe they come from we will be able to see the number of good and bad characters in each universe. We will also be able to compare it to the overall total of good and bad characters included in this superheroes dataset.*

```python
align_counts = superheroes["ALIGN"].value_counts()
print (align_counts)
```

```python
align_counts = superheroes["ALIGN"].value_counts()
#print(align_counts)

align_counts.plot(kind = "bar", fontsize = 15)
plt.title("Number of Good and Bad Characters", fontsize = 18)
plt.show()

align_counts.plot(kind = "pie", fontsize = 14, autopct = "%1.0f%%")
```

```python
#print alignment of character and what universe they come from
align_universe = pd.crosstab(superheroes["ALIGN"],superheroes["UNIVERSE"])
align_universe
align_universe.plot(kind = "bar", fontsize = 15)
plt.title("Number of Good and Bad Characters in each Universe", fontsize = 18)
plt.show()
align_counts.plot(kind = "pie", fontsize = 14, autopct = "%1.0f%%")
```

**Analysis:**
*From the first bar graph and pie chart we can see that there are more bad characters than good characters. Of the total number of characters 41% are bad characters and 32% are good characters. In the second bar graph we can see that the Marvel universe has more good and bad characters than the DC universe. In the Marvel universe there are still more bad characters than good characters just like in the total number of characters bar chart. In the DC universe there are about the same amount of good and bad characters in total. 

I decided to leave the ND and neutral characters so we are able to see that the majority of characters are either good or bad. With this data we can clearly see that the Marvel universe favors bad characters over good characters.*

**Part 2:** *In what ways does the alignment of the characters correspond to the sex of the characters?*

**Description:** *In analyzing the alignment of the characters and their sex, we can see if being a male or female will depend on whether or not a character will be a good one or a bad one. By doing this, we will also be able to see which sex each universe favors more for their good and bad characters.*

```python
#print alignment of character and their sex
align_sex = pd.crosstab(superheroes["ALIGN"],superheroes["SEX"])
align_sex
align_sex.plot(kind = "bar", fontsize = 15)
plt.title("Alignment of Character in Comparison to the Sex of the Character", fontsize = 18)
plt.show()
align_counts.plot(kind = "pie", fontsize = 14, autopct = "%1.0f%%")
```

**Analysis:** *In the bar graph above we can see that the majority of good and bad characters are male. In all the types of sex listed in the bar graph the male characters completely overun the female characters. The green bar representing female characters are significantly lower compared to that of the male characters.* 

*This analyzation allows to see how underepresented females are in both the DC and Marvel comics. With this data we can clearly understand the purpose of FiveThirtyEight's that "Comice book are still made by men, for men and about men".*

**Part 3:** *How does the alignment of characters compare to the ID of the characters?*

**Description:** *By analzying the alignment of characters and their identifications we will be able to see if the type of character alignment dictates whether or not the character will have a secret or public identity.*

```python
#print alignment of character and what universe they come from
align_id = pd.crosstab(superheroes["ALIGN"],superheroes["ID"])
align_id
align_id.plot(kind = "bar", fontsize = 15)
plt.title("How many good & bad characters have secret & public identities?", fontsize = 18)
plt.show()
align_counts.plot(kind = "pie", fontsize = 14, autopct = "%1.0f%%")
```

 **Analysis:** *From the graph labeled "How many good & bad characters have secret & public identities?", we can see that the most characters with secret identities are bad characters. Just a little over 2,000 characters in the bad character column have public identities. Analyzing the good characters column, the majority of the characters have a public identity. Good characters and their secret identity almost comes in close with the public identites but there are still more good characters with public identities than secret ID's. Being that there are more secret identities for bad characters than good characters and vice versa, we can infer that the Marvel and DC comics do this as a way to keep the suspense of whether or not their characters are villians.*
